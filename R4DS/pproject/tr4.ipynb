```r
df1 <- read_csv("d1.csv")
df2 <- read_csv("d2.csv")
df3 <- read_csv("d3.csv")
df4 <- read_csv("d4.csv")
df5 <- read_csv("d5.csv")
```

```r
df1 <- df1 %>% clean_names()
df2 <- df2 %>% clean_names()
df3 <- df3 %>% clean_names()
df4 <- df4 %>% clean_names()
df5 <- df5 %>% clean_names()
```

```r
df1 <- df1 %>% rename(category = company_name, quantity_000_metric_tonnes = quantity_000_metric_tonnes)
df2 <- df2 %>% rename(category = products, quantity_000_metric_tonnes = quantity_000_metric_tonnes)
df3 <- df3 %>% rename(category = oil_companies, quantity_000_metric_tonnes = quantity_000_metric_tonnes)
df4 <- df4 %>% rename(category = products, quantity_000_metric_tonnes = quantity_000_metric_tonnes)
df5 <- df5 %>% rename(category = products, quantity_000_metric_tonnes = quantity_000_metric_tonnes)
```

```r
df1$source_file <- "d1.csv"
df2$source_file <- "d2.csv"
df3$source_file <- "d3.csv"
df4$source_file <- "d4.csv"
df5$source_file <- "d5.csv"
```

```r
all_data <- bind_rows(df1, df2, df3, df4, df5)
```

```r
# Make sure libraries are loaded
library(tidyr)
library(dplyr)
library(ggplot2)

trade_balance <- all_data %>%
  filter(!is.na(trade)) %>%
  
  # --- FIX IS HERE ---
  # Convert the 'trade' column to all uppercase to standardize it
  mutate(trade = toupper(trade)) %>%
  # --- END OF FIX ---

  group_by(year, trade) %>%
  summarize(total_value = sum(value_in_rupees_crore, na.rm = TRUE), .groups = 'drop') %>%
  pivot_wider(names_from = trade, values_from = total_value, values_fill = 0) %>%
  
  # This line will now work correctly because we are guaranteed to have
  # columns named 'IMPORT' and 'EXPORT'.
  mutate(trade_balance_rupees_crore = EXPORT - IMPORT)

print("--- Yearly Trade Balance (Exports - Imports) ---")
print(trade_balance)

# The plotting code remains the same and should work now
ggplot(trade_balance, aes(x = year, y = trade_balance_rupees_crore)) +
  geom_col(aes(fill = ifelse(trade_balance_rupees_crore > 0, "Surplus", "Deficit"))) +
  geom_hline(yintercept = 0) +
  scale_fill_manual(values = c("Surplus" = "steelblue", "Deficit" = "firebrick")) +
  labs(
    title = "Yearly Trade Balance",
    x = "Year",
    y = "Balance (Rupees Crore)",
    fill = "Status"
  )
```

```r
# First, create a 'trade_type' column to label Usage data
plot_data_time <- all_data %>%
  mutate(trade_type = ifelse(is.na(trade), "Usage", toupper(trade))) %>%
  filter(!is.na(month) & !is.na(year)) # Ensure we have date data

# Ensure months are in the correct order
plot_data_time$month <- factor(plot_data_time$month, levels = c(
  "January", "February", "March", "April", "May", "June", 
  "July", "August", "September", "October", "November", "December"
))

# Create a proper date for sorting (approximate to the 1st of the month)
# Filter out any bad month/year data
plot_data_time <- plot_data_time %>%
  filter(!is.na(month)) %>%
  mutate(date = as.Date(paste(year, month, "1"), format="%Y %B %d"))

# Summarize by date and trade type
time_summary <- plot_data_time %>%
  group_by(date, trade_type) %>%
  summarize(total_quantity = sum(quantity_000_metric_tonnes, na.rm = TRUE), .groups = 'drop')

# Plot the lines
ggplot(time_summary, aes(x = date, y = total_quantity, color = trade_type, group = trade_type)) +
  geom_line(linewidth = 1) +
  geom_point() +
  labs(
    title = "Quantity Trends Over Time",
    x = "Date (Year-Month)",
    y = "Total Quantity (000 Metric Tonnes)",
    color = "Trade Type"
  ) +
  theme_minimal()
```

```r
# Filter for data that has a value and is a trade
value_time_summary <- all_data %>%
  filter(!is.na(trade) & !is.na(month) & !is.na(year)) %>%
  mutate(trade_type = toupper(trade)) %>%
  filter(!is.na(month)) %>%
  mutate(date = as.Date(paste(year, month, "1"), format="%Y %B %d")) %>%
  group_by(date, trade_type) %>%
  summarize(total_value = sum(value_in_rupees_crore, na.rm = TRUE), .groups = 'drop')

# Plot the lines
ggplot(value_time_summary, aes(x = date, y = total_value, color = trade_type, group = trade_type)) +
  geom_line(linewidth = 1) +
  labs(
    title = "Trade Value (Price) Trends Over Time",
    x = "Date (Year-Month)",
    y = "Total Value (Rupees Crore)",
    color = "Trade Type"
  ) +
  theme_minimal()
```

```r
# Create the summary for the pie chart
pie_summary <- all_data %>%
  mutate(trade_type = ifelse(is.na(trade), "Usage", toupper(trade))) %>%
  group_by(trade_type) %>%
  summarize(total_quantity = sum(quantity_000_metric_tonnes, na.rm = TRUE)) %>%
  # Calculate percentages
  mutate(percentage = total_quantity / sum(total_quantity) * 100,
         label = paste0(trade_type, "\n", round(percentage, 1), "%"))

# Create the pie chart
ggplot(pie_summary, aes(x = "", y = percentage, fill = trade_type)) +
  geom_col(width = 1) +
  coord_polar(theta = "y") + # This turns the bar chart into a pie chart
  geom_text(aes(label = label), position = position_stack(vjust = 0.5)) +
  labs(
    title = "Overall Composition of Quantity",
    fill = "Trade Type"
  ) +
  theme_void() # Remove background and axes
```
